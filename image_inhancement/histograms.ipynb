```python
#imports
import matplotlib.pyplot as plt
from skimage import io
from skimage import exposure , img_as_float
import numpy as np
```

```python
#load image
real_image =io.imread('E:/Company/KAVOSHGARAN/Virtual Irrigation/Image/Landsat/NIR.tif')
```

```python
real_image
```

```python
#change real image to float image
float_image = img_as_float(real_image)
```

```python
float_image
```

```python
#demonstrate
#To indicate Original image
fig , ax = plt.subplots(nrows=2 , ncols=2 ,figsize=(25,15))
ax[0,0].imshow(real_image , cmap = 'gray')
ax[0,0].set_title('Original Image')
ax[0,0].axis('off')
ax[0,1].hist(real_image.ravel(), bins=256, density=True , histtype='bar',color='blue')
ax[0,1].ticklabel_format(style='plain')
ax[0,1].set_title('Histogram of Original Image')

#To indicate Float image
ax[1,0].imshow(float_image , cmap = 'gray')
ax[1,0].set_title('Float Image')
ax[1,0].axis('off')
ax[1,1].hist(float_image.ravel(), bins=256, density=True , histtype='bar',color='blue')
ax[1,1].ticklabel_format(style='plain')
ax[1,1].set_title('Histogram of Float Image')
plt.savefig("C:/Users/Lenovo/PycharmProjects/Virtual-Irrigation/image_inhancement/images/histogram.png")
plt.show()
```

```python
# Gamma bios 

gamma_correction = exposure.adjust_gamma(float_image,gamma=0.4, gain=1)
```

```python
#demonstrate
#To indicate Float image
fig , ax = plt.subplots(nrows=2 , ncols=2 ,figsize=(25,15))
ax[0,0].imshow(float_image , cmap = 'gray')
ax[0,0].set_title('Float Image')
ax[0,0].axis('off')
ax[0,1].hist(real_image.ravel(), bins=256, density=True , histtype='bar',color='blue')
ax[0,1].ticklabel_format(style='plain')
ax[0,1].set_title('Histogram of Float Image')

#To indicate gamma image
ax[1,0].imshow(gamma_correction , cmap = 'gray')
ax[1,0].set_title('Gamma Image')
ax[1,0].axis('off')
ax[1,1].hist(gamma_correction.ravel(), bins=256, density=True , histtype='bar',color='blue')
ax[1,1].ticklabel_format(style='plain')
ax[1,1].set_title('Histogram of Gamma Image')
plt.savefig("C:/Users/Lenovo/PycharmProjects/Virtual-Irrigation/image_inhancement/images/gamma.png")
plt.show
```

```python
# logarothm adjust bios
lod_adj = exposure.adjust_log(float_image , gain=1 , inv=False )
```

```python
#demonstrate
#To indicate Float image
fig , ax = plt.subplots(nrows=2 , ncols=2 ,figsize=(25,15))
ax[0,0].imshow(float_image , cmap = 'gray')
ax[0,0].set_title('Float Image')
ax[0,0].axis('off')
ax[0,1].hist(real_image.ravel(), bins=256, density=True , histtype='bar',color='blue')
ax[0,1].ticklabel_format(style='plain')
ax[0,1].set_title('Histogram of Float Image')

#To indicate lod_adj image
ax[1,0].imshow(lod_adj , cmap = 'gray')
ax[1,0].set_title('lod_adj Image')
ax[1,0].axis('off')
ax[1,1].hist(lod_adj.ravel(), bins=256, density=True , histtype='bar',color='blue')
ax[1,1].ticklabel_format(style='plain')
ax[1,1].set_title('Histogram of lod_adj Image')
plt.savefig("C:/Users/Lenovo/PycharmProjects/Virtual-Irrigation/image_inhancement/images/lod_adj.png")
plt.show
```

```python
#adjust sigmoid bios
adj_sigmoid = exposure.adjust_sigmoid(float_image ,cutoff= 0.5 ,gain=10 ,inv=False)
```

```python
#demonstrate
#To indicate Float image
fig , ax = plt.subplots(nrows=2 , ncols=2 ,figsize=(25,15))
ax[0,0].imshow(float_image , cmap = 'gray')
ax[0,0].set_title('Float Image')
ax[0,0].axis('off')
ax[0,1].hist(real_image.ravel(), bins=256, density=True , histtype='bar',color='blue')
ax[0,1].ticklabel_format(style='plain')
ax[0,1].set_title('Histogram of Float Image')

#To indicate adj_sigmoid image
ax[1,0].imshow(adj_sigmoid , cmap = 'gray')
ax[1,0].set_title('adj_sigmoid Image')
ax[1,0].axis('off')
ax[1,1].hist(adj_sigmoid.ravel(), bins=256, density=True , histtype='bar',color='blue')
ax[1,1].ticklabel_format(style='plain')
ax[1,1].set_title('Histogram of lod_adj Image')
plt.savefig("C:/Users/Lenovo/PycharmProjects/Virtual-Irrigation/image_inhancement/images/adj_sigmoid.png")
plt.show
```

```python
#equalize adaptive
eq_adaptive = exposure.equalize_adapthist(float_image ,clip_limit= 0.1 ,nbins= 256 )
```

```python
#demonstrate
#To indicate Float image
fig , ax = plt.subplots(nrows=2 , ncols=2 ,figsize=(25,15))
ax[0,0].imshow(float_image , cmap = 'gray')
ax[0,0].set_title('Float Image')
ax[0,0].axis('off')
ax[0,1].hist(real_image.ravel(), bins=256, density=True , histtype='bar',color='blue')
ax[0,1].ticklabel_format(style='plain')
ax[0,1].set_title('Histogram of Float Image')

#To indicate eq_adaptive image
ax[1,0].imshow(eq_adaptive , cmap ='gray')
ax[1,0].set_title('eq_adaptive Image')
ax[1,0].axis('off')
ax[1,1].hist(eq_adaptive.ravel(), bins=256, density=True , histtype='bar',color='blue')
ax[1,1].ticklabel_format(style='plain')
ax[1,1].set_title('Histogram of eq_adaptive Image')
plt.savefig("C:/Users/Lenovo/PycharmProjects/Virtual-Irrigation/image_inhancement/images/eq_adaptive.png")
```
