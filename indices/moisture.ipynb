```python
#imports
import matplotlib.pyplot as plt
import numpy as np 
from matplotlib.pyplot import cm 
from skimage import io
from skimage import exposure
```

```python
#load images
image1 = io.imread('E:/Company/KAVOSHGARAN/Virtual Irrigation/Image/Landsat/Blue.tif')
image2 = io.imread('E:/Company/KAVOSHGARAN/Virtual Irrigation/Image/Landsat/Red.tif')
image3 = io.imread('E:/Company/KAVOSHGARAN/Virtual Irrigation/Image/Landsat/Green.tif')
image4 = io.imread('E:/Company/KAVOSHGARAN/Virtual Irrigation/Image/Landsat/NIR.tif')
image5 = io.imread('E:/Company/KAVOSHGARAN/Virtual Irrigation/Image/Landsat/SWIR1.tif')
image6 = io.imread('E:/Company/KAVOSHGARAN/Virtual Irrigation/Image/Landsat/SWIR2.tif')
```

```python
#changing to array
b1 = np.array(image1).astype(float)
b2 = np.array(image2).astype(float)
b3 = np.array(image3).astype(float)
b4 = np.array(image4).astype(float)
b5 = np.array(image5).astype(float)
b6 = np.array(image6).astype(float)
```

```python
#Normalization
blue = (b1 - np.min(b1))/(np.max(b1)-np.min(b1))
red = (b2 - np.min(b2))/(np.max(b2)-np.min(b2))
green = (b3 - np.min(b3))/(np.max(b3)-np.min(b3))
nir = (b4 - np.min(b4))/(np.max(b4)-np.min(b4))
swir1 = (b5 - np.min(b5))/(np.max(b5)-np.min(b5))
swir2 = (b6 - np.min(b6))/(np.max(b6)-np.min(b6))
```

# INDICES

```python
#NDVI indices
NVDI = (nir - red)/(nir + red)
```

```python
plt.figure(figsize=(25,15))
plt.imshow(NVDI,cmap = cm.Greens)
plt.title('NDVI Image', fontsize = 25)
plt.colorbar()
plt.axis('off')
plt.show
```
