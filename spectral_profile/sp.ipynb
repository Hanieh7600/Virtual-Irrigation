```python
# imports
# import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from skimage import exposure
```

```python
#To create function of even of every pixle with a click
def mouse_click (event , x , y , flags , param):
    if event == cv2.EVENT_LBUTTONDOWN:
        xaxis = []
        yaxis = []
        for i in range (len(image_collocation)):
            #for every band
            xaxis.append('Band{}'.format(i+1))
            #for band's feature
            yaxis.append(image_collocation[i][y,x])
            
        plt.plot(xaxis , yaxis)
        plt.title("spectral profile")
        plt.show()
```

```python
#load images
image_collocation = io.imread_collection('E://Company//KAVOSHGARAN//Virtual Irrigation//Image//Landsat//*.tif')
```

```python
#To get just a band
image = np.array(image_collocation[4])
img = exposure.adjust_log(image)
```

```python
#To show the band (image)
plt.plot()
plt.imshow(img)
plt.show()
```

```python
#create cv2 window to computer our target
cv2.namedWindow('MouseClick',cv2.WINDOW_NORMAL)
cv2.setMouseCallback('MouseClick',mouse_click)

while True:
    cv2.imshow('MouseClick',img)
    if cv2.waitkey(20) == 27:
        break

        
cv2.destroyWindows()
```
