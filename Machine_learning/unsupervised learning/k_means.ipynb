```python
#imports
from sklearn.cluster import KMeans
from skimage import io
import matplotlib.pyplot as plt
```

```python
#load image
image = io.imread('E:/Company/KAVOSHGARAN/Virtual Irrigation/Image/Landsat/NIR.tif')
```

```python
#To get each shape
colms , rows = image.shape[0] , image.shape[1]
```

```python
#reshaping
X = image.reshape((colms*rows),1)
```

```python
X.shape
```

```python
X
```

```python
#creating kmeans
kmeans = KMeans(n_clusters = 4 , random_state = 0)
kmeans.fit(X)
```

```python
#To demonestrate center and lable each cluster
cluster_center = kmeans.cluster_centers_
cluster_lable = kmeans.labels_
```

```python
cluster_center
```

```python
cluster_lable
```

```python
clustered_image = cluster_center[cluster_lable].reshape(colms , rows)
```

```python
clustered_image.shape
```

```python
#demonsterate
plt.figure(figsize=(25,25))
plt.imshow(clustered_image)
plt.title('Clustring', fontsize = 45)
plt.axis('off')
plt.savefig('C:/Users/Lenovo/Desktop/Virtual-Irrigation\Machine_learning/unsupervised learning/images/clustring.png')
plt.show()
```
