```python
#imports
from skimage import io
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
#load images
image_collocation = io.imread_collection('E://Company//KAVOSHGARAN//Virtual Irrigation//Image//Landsat//*.tif', plugin='tifffile')
#To combine all dimensions
images = []
```

```python
#To add every band to images and To flatting images
for i in range(len(image_collocation)):
    images.append(image_collocation[i].flatten())
```

```python
#To demonstrate shape for each band
shape1 = image_collocation[0].shape
```

```python
shape1
```

```python
images
```

```python
#computing PCA and fitting
pca = PCA()
pca.fit(images)
```

```python
#To see components of pca
pca.components_.shape
```

```python
#To see mean of pca
pca.mean_
```

```python
pca1 = pca.components_[0].reshape(shape1)
pca2 = pca.components_[1].reshape(shape1)
pca3 = pca.components_[2].reshape(shape1)
pca4 = pca.components_[3].reshape(shape1)
pca5 = pca.components_[4].reshape(shape1)
pca6 = pca.components_[5].reshape(shape1)
```

```python
pca1.shape
```

```python
#Demonestrate
fig , ax = plt.subplots(nrows = 3 , ncols = 2 , figsize = (20,30))
ax[0,0].imshow(pca1 , cmap = 'gray')
ax[0,0].set_title("PCA_1")
ax[0,0].axis('off')
ax[0,1].hist(pca1.ravel(), bins=256, density=True , histtype='bar',color='blue')              
ax[0,1].set_title('Histogram of PCA_1')
#################################
ax[1,0].imshow(pca2, cmap = 'gray')
ax[1,0].set_title("PCA_2")
ax[1,0].axis('off')
ax[1,1].hist(pca2.ravel(), bins=256, density=True , histtype='bar',color='blue')              
ax[1,1].set_title('Histogram of PCA_2')
#################################
ax[2,0].imshow(pca3 , cmap = 'gray')
ax[2,0].set_title("PCA_3")
ax[2,0].axis('off')
ax[2,1].hist(pca3.ravel(), bins=256, density=True , histtype='bar',color='blue')              
ax[2,1].set_title('Histogram of PCA_3')
plt.savefig('C:/Users/Lenovo/Desktop/Virtual-Irrigation/PCA_algorithm/images/pca.png')
plt.show()
```
