```python
#imports
import matplotlib.pyplot as plt
from skimage.color import rgb2hsv
from skimage import io
import numpy as np
```

```python
#load image 
image_collection = io.imread_collection ("E:/Company/KAVOSHGARAN/Virtual Irrigation/Image/Landsat/*.tif")
```

```python
#combine 3 images
image_stack = np.dstack((image_collection[5],image_collection[4],image_collection[3]))
```

```python
#change rgb to hsv
hsv = rgb2hsv(image_stack)
```

```python
#get hue and saturation and value
hue = hsv[:,:,0]
saturation = hsv[:,:,1]
value = hsv[:,:,2]
```

```python
#Denonestrate
fig , ax = plt.subplots(nrows=2 , ncols=2 ,figsize=(28,28))
#Hue image
ax[0,0].imshow(hue)
ax[0,0].set_title('HUE Image', fontsize =25)
ax[0,0].axis('off')
#Saturation image
ax[0,1].imshow(saturation)
ax[0,1].set_title('Saturation Image', fontsize =25)
ax[0,1].axis('off')
#Value image
ax[1,0].imshow(value)
ax[1,0].set_title('Value Image', fontsize =25)
ax[1,0].axis('off')
#Hsv image
ax[1,1].imshow(hsv)
ax[1,1].set_title('HSV Image', fontsize =25)
ax[1,1].axis('off')
plt.savefig('C:/Users/Lenovo/Desktop/Virtual-Irrigation/hsv/images/hsv_images.png')
```
