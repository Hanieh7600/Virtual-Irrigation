```python
#imports
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
import cartopy.crs as ccrs
```

```python
#load data as read file
data = Dataset('E:/Company/KAVOSHGARAN/Virtual Irrigation/Image/NetCDF4/MERRA2_300.tavgM_2d_slv_Nx.201001.nc4', mode= 'r')
```

```python
data
```

```python
#extraction longtitude and latitude data
lons = data.variables['lon'][:]
lats = data.variables['lat'][:]
T2M = data.variables['T2M'][:,:,]
#To change vector to matrix
lon ,  lat = np.meshgrid(lons , lats)
```

```python
#Demonestrate
fig = plt.figure(figsize=(14 , 7))
ax = plt.axes(projection=ccrs.Robinson())
# ax.set_global()
ax.coastlines(resolution ='110m' , linewidth = 3)
ax.gridlines(linestyle='--' , color = 'black')
plt.savefig('C:/Users/Lenovo/Desktop/Virtual-Irrigation/NC/images/Robinson.png')
```

```python
#Demonestrate
fig = plt.figure(figsize=(14 , 7))
ax = plt.axes(projection=ccrs.Orthographic())
# ax.set_global()
ax.coastlines(resolution ='110m' , linewidth = 3)
ax.gridlines(linestyle='--' , color = 'black')
plt.savefig('C:/Users/Lenovo/Desktop/Virtual-Irrigation/NC/images/Orthographic.png')
```

```python
# Full Demonestrate
fig = plt.figure(figsize=(14 , 7))
ax = plt.axes(projection=ccrs.Robinson())
# ax.set_global()
ax.coastlines(resolution ='110m' , linewidth = 3)
ax.gridlines(linestyle='--' , color = 'black')
##################################################################
celvs = np.arange(230,311,5)
plt.contourf(lon , lat , T2M , celvs , transform = ccrs.PlateCarree() , cmap = plt.cm.jet)
plt .title('MERRA_2 Air Temperture at 2m, januray 210' , fontsize =20)
cb= plt.colorbar(ax = ax , orientation = 'vertical' , pad =0.02 , aspect = 16 , shrink = 0.8)
cb.set_label('K' , size =12 , rotation =0 ,labelpad = 15)
cb.ax.tick_params(labelsize = 10)
plt.savefig('C:/Users/Lenovo/Desktop/Virtual-Irrigation/NC/images/full_Robinson.png')
```
