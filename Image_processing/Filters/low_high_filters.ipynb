```python
#imports
import matplotlib.pyplot as plt
from skimage import io
import cv2
```

```python
#load image
image = io.imread('E:/Company/KAVOSHGARAN/Virtual Irrigation/Image/Geo-Eye/pan_img.tif')
```

```python
#kernel size have to be odd
kernel_size = 3
```

# LOW PASS FILTERS

```python
mean_filter = cv2.blur(image , (kernel_size , kernel_size))
median_filter = cv2.medianBlur(image , kernel_size)
gaussian_filter = cv2.GaussianBlur(image ,(kernel_size , kernel_size),0)
```

# HIGH PASS FILTERS

```python
sobel_filter = cv2.Sobel(image , 0 , dx = 0 , dy = 1 , ksize = kernel_size)
laplacian_filter = cv2.Laplacian(image , -1 , ksize = kernel_size)
```

# Demonestrate

```python
#OriginalImage
plt.figure(figsize=(40,40))
plt.subplot(3,2,1)
plt.imshow(image , cmap = 'gray')
plt.title('Original Image' , fontsize = 40)
plt.axis('off')

#mean_filter
plt.subplot(3,2,2)
plt.imshow(mean_filter , cmap = 'gray')
plt.title('Mean Filter', fontsize = 40)
plt.axis('off')

#median_filter
plt.subplot(3,2,3)
plt.imshow(median_filter , cmap = 'gray')
plt.title('Median Filter', fontsize=40)
plt.axis('off')

#gaussian_filter
plt.subplot(3,2,4)
plt.imshow(gaussian_filter , cmap = 'gray')
plt.title('Gaussian Filter', fontsize=40)
plt.axis('off')

#sobel_filter
plt.subplot(3,2,5)
plt.imshow(sobel_filter , cmap = 'gray')
plt.title('Sobel Filter', fontsize=40)
plt.axis('off')

#laplacian_filter
plt.subplot(3,2,6)
plt.imshow(laplacian_filter , cmap = 'gray')
plt.title('Laplacian Filter', fontsize=40)
plt.axis('off')
plt.savefig('C:/Users/Lenovo/Desktop/Virtual-Irrigation/Image_processing/Filters/images/filters.png')
plt.show()
```
